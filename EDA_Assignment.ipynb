```python
#Que 1. What is the range of selling prices in the dataset.

range_of_selling_price = df["selling_price"].max() - df["selling_price"].min()
print(f"The range of selling price is {range_of_selling_price}")
```

```python
#Que4. How many bikes have driven more than 50,000 kilometers ?

bikes_driven_more_than_50000 = df[df["km_driven"] > 50000].value_counts().sum()
print(f"The number of bikes driven more than 50000 is : {bikes_driven_more_than_50000}")
```

```python
#Que5.What is the average km_driven value for each ownership type ?

average_km_driven_by_ownership = df.groupby("owner")["km_driven"].mean()
print(average_km_driven_by_ownership)
```

```python
#Que7. What is the trend of missing values across the dataset ?

missing_values_trend = df.isnull().sum()
print(f"Missing values trend : \n{missing_values_trend}")
```

```python
#Que8. What is the highest ex_showroom_price recorded, and for which bike ?

highest_ex_showroom_price = df['ex_showroom_price'].max()
bike_with_highest_ex_showroom_price = df[df['ex_showroom_price'] == highest_ex_showroom_price]['name'].values[0]
print(f"The highest ex showroom price recorded is {highest_ex_showroom_price} and for bike {bike_with_highest_ex_showroom_price}")
```

```python
#Que9. What is the total number of bikes listed by each seller type ?

total_bikes_listed_by_seller_type = df.groupby("seller_type")["name"].value_counts().sum()
print(total_bikes_listed_by_seller_type)
```

```python
#Que10. What is the relationship between selling_price and km_driven for first-owner bikes ?

selling_price_km_driven_first_owner = df["selling_price"].corr(df['km_driven'])
print(selling_price_km_driven_first_owner)
```

```python
#Que11. Identify and remove outliers in the km_driven column using the IQR method.

Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
removed_outliers_data = df[(df["km_driven"] >=lower_bound) & (df["km_driven"]<= upper_bound)]
```

```python
print(removed_outliers_data)
print("Dataframe without outliers")
```

```python
#Que12.Perform a bivariate analysis to visualize the relationship between year and selling_price.

bivariate_analysis = df.groupby("year")["selling_price"].mean()
print(bivariate_analysis)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x= "year",y="selling_price",c="r")
plt.title("Bivariate Analysis between year and selling price")
plt.xlabel("Year")
plt.ylabel("selling price")
plt.legend()
plt.show()
```

```python
#Que13. What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?

avg_depreciation_in_selling_price = df.groupby("year")["selling_price"].mean()
print(avg_depreciation_in_selling_price)
```

```python
#Que14. Which bike names are priced significantly above the average price for their manufacturing year

bike_names_above_average_price = df[df["selling_price"] > df.groupby("year")["selling_price"].transform("mean")]
print(bike_names_above_average_price)
```

```python
#Que6.Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist

top_5_most_popular_artists = df.nlargest(5, "Popularity")["Artist"]
average_popularity_by_artist = df[df["Artist"].isin(top_5_most_popular_artists)].groupby("Artist")["Popularity"].mean()
print(f"The highest popularity top 5  artist are:{average_popularity_by_artist}")
print(f"The artist with highest average popularity is : {average_popularity_by_artist.idxmax()}")
plt.figure(figsize=(12,6))
sns.barplot(x=average_popularity_by_artist.index,y=average_popularity_by_artist.values)
plt.title("Average popularity for each artist")
plt.xlabel("Artist")
plt.ylabel("Average Popularity")
plt.show()
```

```python
#Que7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.

# Get the top 5 most popular artists based on total popularity of their tracks
top_5_most_popular_artists = df.groupby("Artist")["Popularity"].sum().nlargest(5).index.tolist()

print("Top 5 most popular artists:", top_5_most_popular_artists)

# For each of the top artists, find their most popular track
most_popular_tracks_for_top_artists = {}
for artist in top_5_most_popular_artists:
  artist_df = df[df["Artist"] == artist]
  if not artist_df.empty:
    most_popular_track = artist_df.nlargest(1, "Popularity")["Track Name"].values[0]
    most_popular_tracks_for_top_artists[artist] = most_popular_track

print("\nMost popular track for each of the top 5 artists:")
for artist, track in most_popular_tracks_for_top_artists.items():
  print(f"- {artist}: {track}")
```
